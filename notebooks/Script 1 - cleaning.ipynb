There are a plenty of boolean columns here counted as int64. Let's convert that for easier visualizations later in the EDA. But first, let's check for inconsistencies with categorical variables.

```python
def show_col_categories(df: pd.DataFrame) -> str:
    df_cols = df.select_dtypes('object').columns.tolist()

    for i, col in enumerate(df_cols):
        print(col, ': ', df[col].unique())
        print()
```

```python
show_col_categories(application_train_df)
```

We will only look into the obvious errors in the dataset.
1. DAYS_EMPLOYED is expecting only negative numbers. However, we will not take action on this right now, and will investigate into this later.
2. Only the flags for owning a car and realty have a Y/N entry, while the others are binary booleans. EMERGENCYSTATE_MODE can also be updated with booleans.

Let's apply that change of elements.

```python
bool_map = {
    'Y': 1,
    'Yes': 1,
    'N': 0,
    'No': 0
}

application_train_df['FLAG_OWN_CAR'] = application_train_df['FLAG_OWN_CAR'].map(lambda x: bool_map.get(x, x))
application_train_df['FLAG_OWN_REALTY'] = application_train_df['FLAG_OWN_REALTY'].map(lambda x: bool_map.get(x, x))
application_train_df['EMERGENCYSTATE_MODE'] = application_train_df['EMERGENCYSTATE_MODE'].map(lambda x: bool_map.get(x, x))
```

For changing int columns to bool:

```python
app_train_df_withbool = application_train_df.copy()

def is_boolean_column(df: pd.DataFrame, col: str) -> bool:
    """
    Args:
        df: pd.DataFrame to check for column
        col: column to check
    Ret:
        bool: identifies int columns that can be turned into bool columns
    """
    # Check which columns have values of only 0 or 1
    if application_train_df[col].dropna().isin([0, 1]).all():
        return True
    return False

# Iteratively check columns containing only 0, 1 or NaN
bool_cols = [col for col in app_train_df_withbool.columns if is_boolean_column(app_train_df_withbool, col)]

# Change datatype of identified columns
app_train_df_withbool[bool_cols] = app_train_df_withbool[bool_cols].astype(bool)
```

## Missing Data

```python
def null_table(df: pd.DataFrame) -> pd.DataFrame:
    """
    This function takes a DataFrame and produces another DataFrame only containing fields with null values.
    Arg: Raw DataFrame, possibly with null values.
    Ret: DataFrame summarizing counts and percentage of null values in each field; fields without null values are removed.
    """
    missing_series = df.isnull().sum()

    # Remove zero-valued null counts
    missing_series_nonzero = missing_series[missing_series.values != 0]

    # Make another series for percent
    pct_missing = missing_series_nonzero / len(df) * 100

    # Combine the two series
    missing_pct_table = pd.concat([missing_series_nonzero, pct_missing], axis = 1)
    missing_pct_table.rename(columns = {0: 'Missing Values', 1: '% Missing'}, inplace=True)

    missing_pct_table.sort_values('Missing Values', ascending=False, inplace=True)

    return missing_pct_table
```

```python
null_table(app_train_df_withbool)
```

There's a lot of missing data in the dataset. Fortunately, the number of records correspond to the count of SK_ID_CURR and the TARGET variable. We'll leave the rest there for now, and decide handling later on in the EDA or Feature Engineering part.

## Check for duplicates

```python
def check_duplicates(df: pd.DataFrame) -> str:
    duplicates = df[df.duplicated()]
    print(f'The dataset has {len(duplicates)} duplicates.')
    print(f'This is {(len(duplicates) / len(df) * 100):.2f}% of the dataset.')
```

```python
check_duplicates(app_train_df_withbool)
```
