We will only look into the obvious errors in the dataset.
1. DAYS_EMPLOYED is expecting only negative numbers. However, we will not take action on this right now, and will investigate into this later.
2. Only the flags for owning a car and realty have a Y/N entry, while the others are binary booleans. EMERGENCYSTATE_MODE can also be updated with booleans.

Let's apply that change of elements.

```python
bool_map = {
    'Y': 1,
    'Yes': 1,
    'N': 0,
    'No': 0
}

application_train_df['FLAG_OWN_CAR'] = application_train_df['FLAG_OWN_CAR'].map(lambda x: bool_map.get(x, x))
application_train_df['FLAG_OWN_REALTY'] = application_train_df['FLAG_OWN_REALTY'].map(lambda x: bool_map.get(x, x))
application_train_df['EMERGENCYSTATE_MODE'] = application_train_df['EMERGENCYSTATE_MODE'].map(lambda x: bool_map.get(x, x))
```

## Missing Data

```python
def null_table(df: pd.DataFrame) -> pd.DataFrame:
    """
    This function takes a DataFrame and produces another DataFrame only containing fields with null values.
    Arg: Raw DataFrame, possibly with null values.
    Ret: DataFrame summarizing counts and percentage of null values in each field; fields without null values are removed.
    """
    missing_series = df.isnull().sum()

    # Remove zero-valued null counts
    missing_series_nonzero = missing_series[missing_series.values != 0]

    # Make another series for percent
    pct_missing = missing_series_nonzero / len(application_train_df) * 100

    # Combine the two series
    missing_pct_table = pd.concat([missing_series_nonzero, pct_missing], axis = 1)
    missing_pct_table.rename(columns = {0: 'Missing Values', 1: '% Missing'}, inplace=True)

    missing_pct_table.sort_values('Missing Values', ascending=False, inplace=True)

    return missing_pct_table
```

```python
null_table(application_train_df)
```

There's a lot of missing data in the dataset. We'll leave it there for now, and decide handling later on in the EDA or Feature Engineering part.
